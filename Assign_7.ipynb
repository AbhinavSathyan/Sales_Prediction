```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas.util.testing as tm
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pickle
```

```python

X_train, X_test, y_train, y_test = train_test_split(df.drop('sales',axis=1), df['sales'], test_size=0.3, random_state=109)
```

```python
X_train
```

```python
linearRegr =  LinearRegression()
linearRegr.fit(X_train, y_train)
y_pred = linearRegr.predict(X_test)
```

```python
#Saving the Model
pickle_out = open("linearRegr1.pkl", "wb") 
pickle.dump(linearRegr, pickle_out) 
pickle_out.close()
```
